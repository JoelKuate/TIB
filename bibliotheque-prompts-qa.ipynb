## Fonctions outils

- `facettes()` — liste les 20 facettes (documents du cycle de test) couvertes par la bibliothèque.
- `chercher(mot_cle)` — cherche un mot-clé dans les prompts (titre, facette, tâche).
- `montrer(id)` — affiche un prompt FN-XX entièrement formaté, prêt à copier.
- `montrer_ce(id)` — affiche un template CE-XX.
- `composer(id_prompt, *id_ce)` — assemble un prompt FN-XX avec un ou plusieurs templates CE-XX en un message unique prêt à coller dans Claude.

```python
def facettes():
    vues = {}
    for pr in PROMPTS:
        vues.setdefault(pr["facette_code"], pr["facette"])
    for code_, nom in vues.items():
        n_prompts = sum(1 for pr in PROMPTS if pr["facette_code"] == code_)
        print(f"{code_:5} {nom:55} ({n_prompts} prompts)")

facettes()
```

```python
def chercher(mot_cle):
    mot_cle = mot_cle.lower()
    resultats = [
        pr for pr in PROMPTS
        if mot_cle in pr["titre"].lower() or mot_cle in pr["facette"].lower() or mot_cle in pr["tache"].lower()
    ]
    print(f"{len(resultats)} résultat(s) pour \"{mot_cle}\" :\n")
    for pr in resultats:
        print(f"  {pr['id']} — {pr['titre']}")
    return resultats

_ = chercher("anomalie")
```

```python
def montrer(id_prompt):
    pr = next((x for x in PROMPTS if x["id"] == id_prompt), None)
    if pr is None:
        print(f"Prompt {id_prompt} introuvable.")
        return
    print(f"{pr['id']} — {pr['titre']}  [{pr['facette']}]")
    print("-" * 70)
    print(f"Rôle : {pr['role']}.")
    print(f"Tâche : {pr['tache']}")
    print(f"Contexte : {pr['contexte']}")
    print(f"Format : {pr['format']}")
    print(f"Contraintes : {pr['contraintes']}")

montrer("FN-14")
```

```python
def montrer_ce(id_ce):
    ce = next((x for x in CE_TEMPLATES if x["id"] == id_ce), None)
    if ce is None:
        print(f"Template {id_ce} introuvable.")
        return
    print(f"{ce['id']} — {ce['nom']}  [{ce['categorie']}]")
    print("-" * 70)
    print(f"Template : {ce['template']}")
    print(f"Pourquoi ça marche : {ce['pourquoi']}")

montrer_ce("CE-19")
```

```python
def composer(id_prompt, *ids_ce):
    pr = next((x for x in PROMPTS if x["id"] == id_prompt), None)
    if pr is None:
        print(f"Prompt {id_prompt} introuvable.")
        return None

    lignes = [f"Rôle : {pr['role']}."]
    lignes.append(f"Tâche : {pr['tache']}")
    lignes.append(f"Contexte : {pr['contexte']}")

    for id_ce in ids_ce:
        ce = next((x for x in CE_TEMPLATES if x["id"] == id_ce), None)
        if ce:
            lignes.append(ce["template"])

    lignes.append(f"Format : {pr['format']}")
    lignes.append(f"Contraintes : {pr['contraintes']}")

    message = "\n".join(lignes)
    print(message)
    return message

# Exemple : le prompt de génération de rapport d'anomalie (FN-14),
# enrichi du référentiel de sévérité (CE-20) et du garde-fou anti-invention (CE-16)
_ = composer("FN-14", "CE-20", "CE-16")
```
